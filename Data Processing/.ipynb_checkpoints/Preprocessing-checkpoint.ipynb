```python
data_df=pd.DataFrame([[np.nan,10,30],[23,np.nan,90],[23,56,89]],columns=list('ABC'))
data_df
```

```python
data_df.dropna()
```

```python
data_df
```

```python
data_df['D']=np.nan
data_df
```

```python
data_df.dropna(how='all',axis=1)
```

```python
data_df.dropna(thresh=1,axis=1)
### Thre is used to remove all null values from the columns
```

```python
data_df.fillna(20)
```

```python
data_df
```

```python
data_df.fillna({'A':20,'B':30,'C':40,'D':23})
```
