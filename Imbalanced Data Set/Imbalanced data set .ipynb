```python
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
lr=LogisticRegression()
grid={'C':10.0**np.arange(-2,5),'penalty':['l1','l2']}
cv=KFold(n_splits=5,)
```

```python
clf=GridSearchCV(lr,grid,cv=cv,scoring='f1_macro')
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(['Class'],axis=1)
y=df.Class
```

```python
train_test_split?
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
class_weight=dict({0:1,1:1000})
rf=RandomForestClassifier(class_weight=class_weight)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# undersampling     is not better
from imblearn.under_sampling import NearMiss
ns=NearMiss(0.8)
X_train_ns,y_train_ns=ns.fit_resample(X_train,y_train)
```

```python
y_train_ns.value_counts()
```

```python
rf=RandomForestClassifier()
rf.fit(X_train_ns,y_train_ns)
y_pred=rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from imblearn.over_sampling import RandomOverSampler   # flase negavite canbe reduced

ns=RandomOverSampler(0.8)
X_train_ns,y_train_ns=ns.fit_resample(X_train,y_train)
```

```python
rf=RandomForestClassifier()
rf.fit(X_train_ns,y_train_ns)
y_pred=rf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# SMOTETomek

from imblearn.combine import SMOTETomek
```

```python
ns=SMOTETomek(0.8)   # new poits are created and oversampling is done
X_train_ns,y_train_ns=ns.fit_resample(X_train,y_train)
```
