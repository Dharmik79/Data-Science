```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('reviews.xlsx')
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
```

```python
 for i in range(len(df)):
        title=re.sub('[^a-zA-Z]',' ',df['Reviews'][i])
        title=title.lower()
        title=title.split()
        title=[ps.stem(word) for word in title if not word in stopwords.words('english')]
        title=' '.join(title)
        corpus.append(title)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_article=CountVectorizer().fit(corpus)
count_tokens=bow_article.get_feature_names()
article_vect=bow_article.transform(corpus)
```

```python
df_count_vect=pd.DataFrame(article_vect.toarray(),columns=count_tokens)
df_count_vect
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf=TfidfVectorizer()
count_df=tf.fit_transform(corpus)
count_features=tf.get_feature_names()
```

```python
df_tfidf=pd.DataFrame(count_df.toarray(),columns=count_features)
df_tfidf.shape
```
