```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
year=[1990,2000,2010,2020]
pop=[2.519,3.45,6.43,2.41]
plt.plot(year,pop)
# plt.plot(x,y)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

# to have y axis lables
plt.yticks([0,2,4,6,8,10])
```

```python
sns.set()
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex',palette='winter_r')
```

```python
sns.barplot(x='total_bill',y='day',data=tips,hue='sex',palette='winter_r')
```

#### color= palette
###### spring,winter_r,coolwarm,husl,autumn

```python
sns.barplot(x='total_bill',y='day',data=tips,hue='sex',palette='spring')
```

###### order attribute is used to use the user defined order


```python
sns.barplot(x='day',y='total_bill',data=tips,order=['Sun','Thur','Sat','Fri'])
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=median)
```

#### the line above the bar graph is known as error and the attribute is ci

```python
sns.barplot(x='day',y='total_bill',data=tips,ci=99)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,ci=40)
```

#### the capsize of the error graph can be controlled using the attribute capsize

```python
sns.barplot(x='day',y='total_bill',data=tips,ci=90,capsize=0.4)
```

```python
sns.barplot(x='size',y='total_bill',data=tips,palette='autumn')
```
