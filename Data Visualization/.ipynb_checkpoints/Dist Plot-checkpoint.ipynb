```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
num=np.random.rand(150)
num
```

```python
sns.distplot(num)
```

```python
label_dist=pd.Series(num,name='variable_x')
sns.distplot(label_dist)
```

### for vertical plot use attribute vertical=True

```python
sns.distplot(label_dist,vertical=True)
```

```python
sns.distplot(label_dist,hist=False)
```
