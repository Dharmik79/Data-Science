```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']=(10,6)
```

```python
df=pd.read_csv('Data/gas_prices.csv')
df.head()
```

```python
plt.plot(df.Year,df.USA,'b.-',label="USA")
plt.plot(df.Year,df.Canada,'g.-',label="Canada")
plt.legend()
plt.xticks(df.Year[::3])
plt.show()
```

```python
for country in df:
    if country != 'Year':
        plt.plot(df.Year,df[country],marker='.',label=country)
plt.legend()
plt.xticks(df.Year[::3])

plt.savefig('Gas Price.png',dpi=300)

plt.show()
```

```python
fifa=pd.read_csv('Data/fifa.csv')
fifa.head()
```

```python
fifa.drop(['Unnamed: 0'],axis=1,inplace=True)
fifa.head()
```

### Plotting Histogram


```python
bins=[50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,rwidth=0.6)
plt.xticks(bins)
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
```

```python
plt.pie([left,right])
```

```python
labels=['left','right']
colors=['red','green']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%',pctdistance=0.6,explode=(0.1,0.01))
```
